# Creating Data

```python
import pandas as pd
```

```python
pd.DataFrame({'Yes': [50,21], 'No': [131, 2]})  # DataFrame with integers
```

```python
# DataFrame with strings

pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 'Sue': ['Pretty good.','Bland.']})
```

```python
# We can assign an index parameter in our DataFrame

pd.DataFrame({'Bob': ['I liked it.', 'It was awful.'], 
              'Sue': ['Pretty good.','Bland.']},
            index =['Product A', 'Product B'])
```

```python
# We can create a Series (it is a list).

pd.Series([1, 2, 3, 4, 5])
```

```python
# We can assign an index parameter to the Series:
pd.Series([30, 35, 40], index=['2015 Sales', '2016 Sales', '2017 Sales'], name='Product A')
```

```python
# A DataFrame can be seen as a bunch of Series.
```

# Reading common file formats

```python
# read_csv() function for reading the Data into a DataFrame.

wine_reviews = pd.read_csv("winemag-data-130k-v2.csv")
```

```python
# shape attribute to check how large the DataFrame is:

wine_reviews.shape
```

```python
# head() command grabs the first five rows:

wine_reviews.head()
```

```python
# We want pandas use the in-built index, so we have to specify index_col

wine_reviews = pd.read_csv("winemag-data-130k-v2.csv", index_col=0)
wine_reviews.head()
```

```python
# We want to read an excel file: we have to specify the name of the sheet of interest.

wic = pd.read_excel("WICAgencies2013ytd.xls", sheet_name='Total Women')
wic.head()
```

```python
wic_pregnant = pd.read_excel("WICAgencies2013ytd.xls", sheet_name='Pregnant Women Participating')
wic_pregnant.head()
```

```python
# We can read also SQL files (where most of the data on the web gets stored). 
# You need to create a connector. 

import sqlite3
conn = sqlite3.connect("FPA_FOD_20170508.sqlite")
```

```python
# You need also to write a SQL statement. We need to use SQL to load data  into.
# Every SQL statement begin with SELECT. 
# The name of the table is fires.

fires = pd.read_sql_query("SELECT * FROM fires", conn)
```

```python
fires.head()
```

# Writing common file formats

```python
# To write a csv file

wine_reviews.head().to_csv("wine_reviews.csv")
```

```python
# To write an excel file we need the sheet_name:

wic.to_excel('wic.xlsx', sheet_name='Total Women')
```

```python
# To write a SQL database, we need a connector

conn = sqlite3.connect("fires.sqlite")
fires.head(10).to_sql("fires", conn)
```
